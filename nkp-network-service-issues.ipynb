######  2. Incorrect Gateway or Netmask
***

After completing this topic, you should be able to: ``correct issues relating to incorrect gateway or netmask``

Every system that uses TCP/IP as its communication protocol requires three configured values. If you want to be able to communicate with systems on your own subnet, as well as those outside of your subnet, or perhaps, most notably, the Internet. You need an IP address, a subnet mask, and a default gateway. So in this presentation, we will take a look at some issues that can arise from incorrectly configured systems. Now beginning with the default gateway. This enables the local network, and all the hosts on that network, to connect to an external network such as the Internet. And a router is commonly used as a gateway. So the physical device itself is a router, because they are responsible for connecting networks together. But it represents the means for us to leave our network and get to any other network, so we refer to it often as a gateway. But you will hear those two terms used quite interchangeably. And this is the same for a home environment as it is for any kind of enterprise environment. To get to another network, you have to go through a router. Now, you can also use a computer with two network adapters for a router or your gateway. As long as it has one adapter that is connected to the local subnet and another one that is connected to the outside network such as the Internet. In fact, just about every version of Windows Server has the ability to act as a router. So you don't need to go out and purchase an official router, for lack of a better word, but it really doesn't matter, as long as it does the job. And as referred to as the default gateway, simply because any given device or any given application will use the gateway unless another one is specified. But it will also forward the packet to the default gateway, if it is deemed that the intended recipient is not within your subnet. And that's actually the job of the subnet mask. It compares the destination to the subnet mask. And, in essence, says that this packet is either destined for a host within the subnet or it's in some other network. And if it's in some other network, then it will default to forward the packet off to the gateway. Now you can in fact have more than one default gateway defined, but in most cases, I think you'll see that there is only one. And, in essence, that represents the way to get out of this network. Now if you are unable to reach the default gateway, then of course you need to isolate the problem. So begin with checking your own local IP configuration and of course the physical cabling if the problem is only affecting one host. If it's a problem with the router itself, which of course it could be, then it's likely affecting the entire subnet. So of course try to determine what the scope of the problem is. Now the subnet mask or often referred to as just the netmask, is a 32-bit mask that's used to divide an IP address. Every IP address is divided into two sections, the network portion which identifies which network you are on. And then the host portion which identifies the specific host on that network. So divides the network up into smaller subnets and then determines how many hosts can be on that subnet. Now this is something we discussed in detail way back in course 2, so if you aren't familiar with this, please go back and review that course. [A table is displayed, showing the division of an IP address. There are three rows: Netmask, Binary, and Netmask length, and four columns of numbers. Netmask has columns: 225., 225., 225., and 225. Binary has four colums of 11111111, and Netmask length has columns 8, 16, 24, and 32. ] But as a quick summary, if we see the binary values of 1s, and if they were all set to 1, the decimal value would equal 255. And then the length is quite simply how many of those bits are set to 1. Now you would not see all 1s across the board in any practical implementation of subnetting. So here we see the examples that are the default subnet masks for the class of the network, Class A, B, or C. [A table is displayed of the default subnet classes with five columns: Class, Netmask length, # of networks, # of hosts, and Netmask. There are three Classes, Class A, B, and C. Class A has a Netmask length of 8,126 # of networks, 16,777,214 # of hosts, and a 255.0.0.0 Netmask. Class B has a Netmask length of 16, 16,382 # of networks, 65,534 # of hosts, and a 255.255.0.0 Netmask. Class C has a Netmask length of 24, 2,097,150 # of networks, 254 # of hosts, and a 255.255.255.0 Netmask. ] And the class simply defines the number of networks as well as how large they are in terms of how many hosts can be accommodated on that particular network. So you can see that Class A networks are very, very large. They can contain millions of hosts, but there aren't very many of them. So the class B is a little more reasonable. There are 16,000 class B networks, and they're fairly large, 65,000 hosts. Class C's, there are million of them but they are all very small. But those are just the defaults. And again, you see the length of the mask is either 8, 16 or 24 and that corresponds with the decimal values we see under the netmask column. 255 is 8 bits enabled, the remaining 24 are off. Class B is split right down the middle. 16 bits for the netmask length equates to 255.255.0.0. That's 16 bits on, 16 off. And then Class C is 24 bits on, 8 bits off for 255.255.255.0. But again, those are just starting points. You can use different subnet masks to vary the size to whatever you need. So when troubleshooting the subnet mask, we refer to any documentation that outlines what subnet a particular host should be on. Check the other devices and confirm that they are using the same subnet mask as you. They should all match. All systems in the same subnet should have the same value for the subnet mask. And this is certainly something that can happen if someone manually configured any particular host. Because a simple typo could make that host believe that they are in fact on a different subnet. Now some of the tools you have to assist you in troubleshooting include some command line utilities, most notably Ipconfig and/or Ping. Ipconfig will simply show you what values have been configured. And Ping can be used to test connectivity with other systems on your subnet, or of course the default gateway. So you just ping its address, and if you get a response, you are able to see that system. And if you don't get a response, then clearly something is wrong. These tools are installed by default on most operating systems. And they can be used to determine how traffic is being directed throughout the network. So how far is it getting and is it being prevented from being routed externally? In other words, if I can ping the interface of the router that faces me but I can't seem to ping anything beyond that, then the problem likely is external. But that still helps you to isolate where the problem is. So again, using a combination of these tools and approaches, you can hopefully isolate what the problem is. But ultimately, you must have a valid IP address, a correct subnet mask and a default gateway that is up and running, to be able to fully internetwork with TCP/IP.

######  3. Duplicate IP or MAC Addresses
***

After completing this topic, you should be able to: ``troubleshoot issues caused by duplicate IP and MAC addresses``

In this presentation, we'll take a look at some problems you may encounter if duplicate addresses are discovered in a network. And we'll begin with the IP address. This occurs when the same IP address is assigned to two network devices within the same network. In that event, one or both devices can end up being disabled or, of course, just losing connectivity. Now, this can happen for a number of reasons. Typically it's some kind of configuration error, particularly if the same static IP address was used twice. If you have a number of systems in your environment that are statically configured, then you need to ensure that all of the values are unique. And that could just be a simple typo, for example. But an IP address in a DHCP range might still just be handed out to a client. But perhaps the address it hands out is already being statically used by some other system. And many environments, the servers, for example, might be statically configured. So if that's the case, the DHCP scope that is configured has to exclude the values that are being statically used. The DHCP service itself might encounter an error. That's not as likely, but it is possible. And perhaps it could also be an issue of a system returning online, something that has been turned off for quite some time, and it had a particular address that has now been assigned elsewhere. So when that system comes back online, it still has that old address and, of course, now you have the duplicate issue. Now this is certainly more of an issue if you are using a lot of static IP addresses in your environment. If you have many administrators and many systems that are being configured statically, of course, you are more likely to encounter the same address being assigned twice. So it just becomes important to document the allocation. And, of course, to verify which systems are using which addresses and just ensure that there aren't any duplications. Static addresses do not change. So once it's set, it stays that way forever. So, again, just verify that it is correct, and that it is unique. Now this does involve a little more administrative overhead to configure these IP addresses statically. So it's not ideal for larger networks. But it can be a little more stable and a little more reliable because any system that does have a static address no longer has to rely on a DHCP server to assign it its address. So if you know that any given system is basically always going to stay put in terms of its network location and just is always going to use that address, by all means, configure it manually. Now you have fewer systems to deal with in terms of DHCP. But again, you just have to ensure that you do not assign duplicate addresses statically. And whichever addresses are being used statically must be exempted from the DHCP server scope. So in terms of troubleshooting duplicate addresses, certainly review the documentation. Ping the address first. If it responds, then, of course, it is already in use. Perhaps refer to MAC address tables. Switches maintain MAC tables whereby it knows which device is using any given IP address. And you can also do some packet capturing if you really need to be detailed. But in most cases, at least with Windows systems, they're pretty good at detecting it. As soon as you boot up a system that does cause a conflict, it should actually tell you that it is detecting this address elsewhere in the environment. But unfortunately it does not say which system is using it. Now, duplicate MAC addresses are far less common, but they can occur. And the MAC address refers to the 12-digit hexadecimal number, which is 48 bits in length. That is sourced directly from the hardware, the network interface card itself has this address burned into it. And the address is globally unique and assigned by the manufacturer. So it's not very common to have manufacturers produce network cards with duplicate addresses. But it has happened in the past because they think an address that was used quite some time ago is now defunct. So it's possible, but very unlikely. Generally, duplicate MAC addresses occur when the vendor-assigned address is not used. In other words, it's replaced with a self-assigned address or what's referred to as a locally administered address. Now again, this is not particularly common either but there are certain adapters that allow you to effectively change what MAC address is going to be used for that device. Now, you can find out what address is being used with the ARP utility. You can view or modify the ARP cache. And the entries include the MAC addresses. And it basically says that this MAC address is currently using this IP address. And the ARP utility does allow you to view and/or modify those values. So you can use this to troubleshoot problems that are related to duplicate IP addresses or MAC addresses. But again, in most cases, almost all network administrators are generally fine with using the burned in MAC address. But if it is a case, for example, where a server has a faulty network adapter card, and you replace the adapter card, then you do have to assign the IP address that was being used for that system to the same adapter. But now there is a different MAC address. So you do have to update some of the values, possibly, in some ARP caches to ensure that they are aware of the new MAC address using that old IP address. Now, again, that might not be an issue because ARP caches don't remain current for all that long. I believe it's only about two minutes, by default. So if the cache is simply left to purge out, then ultimately it will update itself anyway. But it can be an issue if you have these very important systems that need to be able to communicate via the MAC address and the IP address. So changes can result in problems. But again, issues with duplicate MAC addresses are far less common than issues with duplicate IP addresses. But always something to be on the lookout for.

######  4. Resolving Names
***

After completing this topic, you should be able to: ``troubleshoot issues where names are not resolving``

Now in this presentation, we'll take a look at troubleshooting DNS or the Domain Name System. And this is the service used to translate domain names into IP addresses. And as such DNS is relied upon heavily for Internet browsing and email services. Because of course we, as users, generally do not know the IP addresses of the systems we want to reach, but we do know their names. So DNS issues can generally result in the ability for users to connect to websites or Internet hosts. And often browsers will report No Internet Access if the DNS service is unavailable. Which can, in some cases, lead you to believe that it might be a problem with the default gateway because it seems like we have no Internet. But if you do know the actual IP address of even one Internet site, you can enter that into your browser. And if this site comes up, then clearly it's not a problem with the router but rather with the DNS service itself. Now the NSLOOKUP utility can be used for troubleshooting your internal DNS issues. In essence, trying to determine if the DNS server is responding to your request and/or specifically, which server is responding. Now load balancing refers to having multiple servers and distributing the workload across those servers. But in those cases, it can be difficult to know which server has failed. And this can result in intermittent connectivity, because you might hit a server that is perfectly fine one time. And then you hit a server that is down, the next, so you get this intermittent connectivity. And there can also be issues with DNS forwarders. Any given DNS server has its database, but it can also be configured to forward a request off to a different server if it does not have the record that you are requesting. So if a particular domain, for example, is continuously being requested, and the server you are hitting is configured to forward all of those requests to a specific server that has gone down, then all of them will fail. Now, alternate DNS refers to the fact that most organizations have a primary and a secondary DNS server. So it's similar to load balancing in that there are two servers but the secondary is never used unless the primary goes down or fails. So if you suspect that's the problem and you do switch over to the secondary, you can confirm if the problem relates to the primary server. So again, it's just an alternate, a standby, if you will. And of course, you also want to rule out malware or attacks. There could be a DNS poisoning attack that alters the records. And of course, there could be any other kind of malware that might affect the service itself, such as a denial of service attack or a virus or just anything that brings down the service. So when trying to determine what the problem is, of course, verify the network connectivity. In other words, is the DNS server itself online? Can you see it? So ping it to determine if it is even alive and responding to any kind of request. And then, of course, determine who is being affected. Is it just one user or is it the entire subnet? In terms of some of the tools that you can use, Ping, of course, allows you to simply test, is the server responding. But what you would want to do is ping it by both its IP address and its name because the DNS server itself should be able to resolve its own name. So if it works by the IP address but not by the name, then there is a problem with the service. And again, the NSLOOKUP utility can test the name resolution. It returns the name and the IP address of the DNS server. So you can find out if it's the correct server and of course what its actual address is just in case you have the IP address set wrong in your own system. If your system believes that it should be contacting IP address x but the DNS server is, in fact, IP address Y, then it's not a problem with the server, it's a problem with your own configuration. But ultimately, DNS is, again, pretty much required these days for Internet connectivity, and even just being able to resolve names within your own internal environment. So it's a very important service and something you always want to keep on top of in terms of monitoring its health and status.

######  5. Expired DHCP Scopes and  Rogue DHCP Servers
***

After completing this topic, you should be able to: ``troubleshoot issues caused by expired DHCP scopes and rogue DHCP servers``

In this presentation, we'll take a look at troubleshooting some DHCP issues, beginning with the scope itself. And this is something that you configure on the server. And it represents the valid range of IP addresses that are available for assignment. And any one address from that range is leased to the client computers for a certain amount of time. And that is configurable as well, and we'll talk about that in a moment in a little greater detail. But the scopes are configured to designate a pool of IP addresses. The server can then allocate to the DHCP clients. Now, the clients are required to be configured to use DHCP. In other words, you have a choice. You can set the client to use a static configuration or you can simply tell it to obtain an IP address automatically. In which case you are telling it to request its configuration from a DHCP server. Now this in fact is the default on Windows systems, but it certainly could have been changed. So with respect to the scope itself, there are several properties. These are only a few examples, but it does of course have to indicate the network ID and the subnet mask that you are going to use. Of course the actual IP address range, the starting address through to the ending address, the lease duration. And this is typically around a week by default for most systems. But you can set it to really just about any value you want. And it typically comes down to how many clients you have versus how many addresses you want to configure within a scope. And the idea being that if you have only a few addresses but a lot of systems that maybe aren't all on at the same time, such as shift work, then you would typically have a fairly short lease duration. If you have the exact opposite, a lot of addresses and not that many clients, then any one client can keep that configuration for a fair amount of time. So again, the duration is your call but seven, eight days is usually around the default. The name of the scope is simply what you can call it. And an exclusion range is a set of addresses that falls within the scope, but that should not be assigned. So let's just say, your scope is 192.168.1.100 to 200, normally, it would assign out every one of those addresses from 100 to 200. But if you know, for example, that 150 to 160 is already being used because they've been statically assigned to maybe ten servers, we'll say then that represents the exclusion range. So again, it's really just only values that would normally fall within the range that you just don't want to assign. So when troubleshooting scopes, some common issues can include the scope itself not being activated. This is generally an option you can configure when you are setting things up initially. For example, you might have many scopes that you need to create, but you don't have to activate them right then and there. As soon as it is activated, the server will start allocating those addresses, but maybe you just don't want it to do so yet. So you can leave them in a deactivated state until everything is fully configured. But of course, if it was supposed to be activated and you just forgot, well then clearly that's a problem. The lease configuration, there could be really anything wrong with the lease in terms of the duration and/or the validity periods. Maybe you just made a mistake there. A limited scope typically refers to a scope that does not have enough addresses to assign all of the clients. So you just made it too small basically. And the DHCP service itself might have been stopped. Any given service on any given server could fail at any point, so if that service does fail, then that server will stop allocating addresses. And perhaps something has simply been misconfigured on the client side. Same problem could happen, the service on the client is also there, the DHCP client service may have failed. Maybe somebody set a particular client system to be statically configured and they weren't supposed to, something along those lines. So there can certainly be a lot of issues. Now, you might also encounter a rogue DHCP server, which is simply an unauthorized DHCP server appearing on a network, one that is not under administrative control. Now, this is often used in man-in-the-middle attacks, because every version of Windows Server, for example, allows you to create a DHCP scope. So an attacker could very easily create a new DHCP server and then start handing out bogus IP address configurations to redirect clients. It offers the same services as an authorized DHCP server, so it will absolutely supply you with an IP address, a default gateway, and a DNS server. But of course, if it is an unauthorized server doing so, that address configuration would be incorrect. Now, you can try to detect rogue DHCP servers by using the DHCPLOC or locate utility. This is a free download available from Microsoft. Which is quite simply used to detect the presence of an unauthorized DHCP server. And of course, you can also use the ipconfig /all command on any given client to confirm the IP address of the DHCP server. Because, of course, a rogue DHCP server would not have the same IP address as an authorized server. So in terms of mitigating rogue DHCP servers, you can implement intrusion detection systems. This does require the appropriate signature to detect DHCP packets. You can use multi-layer switches which are capable of dropping packets that aren't authorized. And you can also implement DHCP snooping, which also drops DHCP messages that are not trusted. And if you are in a Windows and Active Directory environment. Active Directory itself, authorizes DHCP servers but this, of course, would only be applicable to Windows/Active Directory client systems. They are then informed to not accept an IP address configuration from any server that is not authorized. But again, that's only applicable to Windows and Active Directory environments. So anything other than that would not be able to detect, of course, the presence of an Active Directory authorized DHCP server. But ultimately, it's something that you want to keep on top of, and ensure that your clients are getting the correct information from the correct servers. And if there are problems, then of course, as mentioned, double-check the address of the server. Double-check the status of the service itself, and of course, the configuration that you are receiving from that server.

######  6. Expired IP Addresses and Incorrect Times
***

After completing this topic, you should be able to: ``correct issues caused by expired ip addresses and incorrect system times``

Now in this presentation, we'll take a look at how time can affect the configuration of a system, particularly with respect to DHCP, as well as some other components. But when it comes to your IP address configuration that you receive from a DHCP server, there is always what's known as a lease duration on that configuration. Meaning that you have it for a certain amount of time before it has to be renewed. And this is applicable to basically any type of device, whether it's a laptop, a desktop, a smartphone, even a smart TV these days, or a tablet. Now, those types of devices toward the bottom there, the smartphones, the smart TVs, and tablets, certainly much more common in a home environment. But it's the same idea, that any time you get an IP configuration from a DHCP server, there is this time associated with the lease. So in terms of the optimum length of what that time should be, it really is up to you. But in an environment such as a small home network, typically there are more addresses than devices. So in that case, you can have a longer lease period because it's unlikely that you will run out of addresses. Most home routers, for example, default to have about a 100 addresses available, and it's usually something in the 192.168., let's just say 1.100 to 200 range. That's fairly common. So that represents 100 available addresses. So the chance that you would have more than 100 devices leasing IP addresses from that server in a home environment is fairly low. But if you are in a coffee shop, for example, then you might have many guests using your Wi-Fi, but only for a short period of time. So in this case, a shorter lease period would make much more sense. You don't want someone to lease an address for, let's say, a week, when they are only going to be there for half an hour. So you can adjust the length of the lease quite a lot. But it ultimately does come down to how long you feel is appropriate for any given device to have the same address configuration. Now, in a corporate environment, again, the default is usually around a week for most devices to have a valid address configuration. But again, if you have far more addresses than devices, you can look at increasing that validity period if you want to. Or if you have the exact opposite, more devices, not that many addresses, you can still shorten it. It really comes down to whatever you feel is appropriate. Now, an expired IP address is essentially the fact that your lease does expire. So you can lose the network configuration for any given network. Now, this might interfere with other devices or services if a device continues to use an IP address from an expired lease. So, again, the idea is if you did receive a particular configuration from a DHCP server and the lease says it's x amount of days, then, of course, at that point in time, when the expiry comes, that configuration is no longer valid. If your system is turned on and is able to contact the DHCP server, it will simply request a new configuration and you shouldn't have too much trouble at that point. But of course, you might lapse over the expiry time with your system turned off. Maybe you know, for example, that the lease configuration is going to expire, maybe let's say at the end of today. But if you shut down your system and then leave for a long weekend or something like that, then that address is still attached to your device the entire time it's off. And then when you boot it up, let's say, maybe over the weekend if you're at home because it was a laptop, then it's possible it will still try to use that expired lease. And it might interfere with trying to get an address in your home network or something along those lines. In the meantime, the server itself might attempt to lease the same IP address to another client on the network because it believes that address, since it did expire, in essence, is now available. So it might lease it out to someone else. So then you come back to work on Monday morning and you boot up your system and it still has that old address, then you might end up with an IP address conflict. So again, it depends on the duration and the state of the system when that expiry happens. So these aren't particularly common scenarios, but they can arise. Now, another issue is the time on the system itself. It might be incorrect. Computer clocks can be inaccurate, and this applies to servers, workstations, and any other network device that maintains the time. And in some cases, they can still be configured manually. And this effectively refers to the fact that the clock can be maintained by a battery, which can drift slightly on any given day. If the battery is getting low, for example, very often you will notice the time starting to lag behind a little bit, and that can have an effect. Because time synchronization is critical for logging, managing, securing, planning, and debugging. And can also cause network issues, such as the inability to join an Active Directory domain, and various other synchronization problems. And this comes down to the fact that clients, in essence, need to agree on what time it is as per their servers. The Kerberos protocol, for example, in a Windows Active Directory environment only has a tolerance of five minutes. If they do not agree that the time is within that five minutes, and that's only by default I should mention, then it will not let that client participate. Because it can throw off logging and managing and securing those systems. Particularly, when it comes to security, if the time is way off, then whatever it did log in terms of security just does not represent the correct time. And that can also have an effect on your DHCP lease if the time is not correct. So you can also implement things such as the Network Time Protocol to try to prevent this. This can be used to synchronize hundreds of millions of devices and computers. And there are also public Internet time servers, you can see the address there. [The address is http://www.pool.ntp.org/] But this allows you to synchronize with these, for lack of a better word, universal time servers, so that they do agree on what time it is. That's, in essence, the purpose of these protocols, so that everyone is synchronized to the same time. This way you know that your logs, and your security settings, and your policies, and your DHCP leases are all operating on systems that agree on what time it is.

######  7. SSL Certificates and Blocked Ports
***

After completing this topic, you should be able to: ``troubleshoot issues caused by untrusted SSL certificates and blocked tcp/udp ports``

In this presentation, we'll talk about some issues that you might encounter with SSL certificates and open ports. Now SSL is the Secure Sockets Layer, and this is a standard security protocol that protects information transmitted online.

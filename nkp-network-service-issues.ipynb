###### Incorrect Gateway or Netmask
***

After completing this topic, should be able to: ``correct issues relating to incorrect gateway or netmask``.

Every system that uses TCP/IP as its communication protocol requires three configured values, if you want to be able to communicate with systems on your own subnet, as well as those outside of your subnet, or perhaps, most notably, the Internet:


- IP address


- Subnet mask


- Default gateway 


Looking at some issues that can arise from incorrectly configured systems.

<a id="DefaultGateway"></a>
***
###### Default Gateway
***

Enables the local network, and all the hosts on that network, to connect to an external network such as the Internet.

A router is commonly used as a gateway - physical device itself is a router - because they are responsible for connecting networks together but it represents the means for us to leave our network and get to any other network, refer to it often as a gateway.

The two terms - gateway / router - used interchangeably and this is the same for a home environment as it is for any kind of enterprise environment - to get to another network, you have to go through a router. 

Can also use a computer with two network adapters for a router or your gateway, as long as it has one adapter that is connected to the local subnet and another one that is connected to the outside network such as the Internet - in fact, just about every version of Windows Server has the ability to act as a router, so you don't need to go out and purchase an official router, for lack of a better word, but it really doesn't matter, as long as it does the job.

Referred to as the default gateway, simply because any given device or any given application will use the gateway unless another one is specified, but it will also forward the packet to the default gateway.

If it is deemed that the intended recipient is not within your subnet and that's actually the job of the subnet mask - it compares the destination to the subnet mask and, in essence, says that this packet is either destined for a host within the subnet or it's in some other network and if it's in some other network, then it will default to forwarding the packet off to the gateway. 

Can in fact have more than one default gateway defined, but in most cases, will see that there is only one and, in essence, that represents the way to get out of this network.

###### Unreachable Default Gateway
***

If you are unable to reach the default gateway, then of course you need to isolate the problem, so begin with checking your own local IP configuration and of course the physical cabling if the problem is only affecting one host. 

If it's a problem with the router itself, which of course it could be, then it's likely affecting the entire subnet, so of course try to determine what the scope of the problem is. 

The subnet mask or often referred to as just the netmask, is a 32-bit mask that's used to divide an IP address - every IP address is divided into two sections, the network portion which identifies which network you are on and then the host portion which identifies the specific host on that network. 

Divides the network up into smaller subnets and then determines how many hosts can be on that subnet. 

See the binary values of ``1`` - if all set to ``1`` - the decimal value would equal ``255`` - the length is quite simply how many of those bits are set to ``1``. 

Would not see all ``1`` across the board in any practical implementation of subnetting.

###### Troubleshooting Netmask
***

So when troubleshooting the subnet mask, we refer to any documentation that outlines what subnet a particular host should be on. 

Check the other devices and confirm that they are using the same subnet mask as you - should all match. 

All systems in the same subnet should have the same value for the subnet mask and this is certainly something that can happen if someone manually configured any particular host because a simple typo could make that host believe that they are in fact on a different subnet. 

Some of the tools you have to assist you in troubleshooting include some command line utilities, most notably ``ipconfig`` and/or ``ping``. 


- ``ipconfig`` will simply show you what values have been configured 


- ``ping`` can be used to test connectivity with other systems on your subnet, or of course the default gateway 


So you just ``ping`` its address, and if you get a response, you are able to see that system and if you don't get a response, then clearly something is wrong. 

These tools are installed by default on most operating systems and they can be used to determine how traffic is being directed throughout the network, so how far is it getting and is it being prevented from being routed externally? 

In other words, if I can ``ping`` the interface of the router that faces me but I can't seem to ``ping`` anything beyond that, then the problem likely is external but that still helps you to isolate where the problem is. 

So again, using a combination of these tools and approaches, you can hopefully isolate what the problem is but ultimately, you must have a valid IP address, a correct subnet mask and a default gateway that is up and running, to be able to fully internetwork with TCP/IP.

<a id="DuplicateIP"></a>
***
###### Duplicate IP
***

This issue occurs when the same IP address is assigned to two network devices within the same network. 

In that event, one or both devices can end up being disabled or, of course, just losing connectivity. 

Can happen for a number of reasons, typically it's some kind of configuration error, particularly if the same static IP address was used twice. 

If you have a number of systems in your environment that are statically configured, then you need to ensure that all of the values are unique and that could just be a simple typo, for example, but an IP address in a DHCP range might still just be handed out to a client but perhaps the address it hands out is already being statically used by some other system. 

In many environments, the servers, for example, might be statically configured, so if that's the case, the DHCP scope that is configured has to exclude the values that are being statically used. 

The DHCP service itself might encounter an error, that's not as likely, but it is possible.

Perhaps it could also be an issue of a system returning online, something that has been turned off for quite some time, and it had a particular address that has now been assigned elsewhere, so when that system comes back online, it still has that old address and, of course, now you have the duplicate issue. 

This is certainly more of an issue if you are using a lot of static IP addresses in your environment - if you have many administrators and many systems that are being configured statically, of course, you are more likely to encounter the same address being assigned twice, so it just becomes important to document the allocation and, of course, to verify which systems are using which addresses and just ensure that there aren't any duplications. 

Static addresses do not change, so once it's set, it stays that way forever, so, again, just verify that it is correct, and that it is unique. 

Now this does involve a little more administrative overhead to configure these IP addresses statically, so it's not ideal for larger networks but it can be a little more stable and a little more reliable because any system that does have a static address no longer has to rely on a DHCP server to assign it its address, so if you know that any given system is basically always going to stay put in terms of its network location and is always going to use that address, by all means, configure it manually - now you have fewer systems to deal with in terms of DHCP but again, you just have to ensure that you do not assign duplicate addresses statically and whichever addresses are being used statically must be exempted from the DHCP server scope.

In terms of troubleshooting duplicate addresses, certainly review the documentation.

``ping`` the address first - if it responds, then, of course, it is already in use. 

Perhaps refer to MAC address tables - switches maintain MAC tables whereby it knows which device is using any given IP address and you can also do some packet capturing if you really need to be detailed but in most cases, at least with Windows systems, they're pretty good at detecting it, as soon as you boot up a system that does cause a conflict, it should actually tell you that it is detecting this address elsewhere in the environment but unfortunately it does not say which system is using it.

<a id="DuplicateMAC"></a>
***
###### Duplicate MAC
***

Duplicate MAC addresses are far less common, but they can occur - the MAC address refers to the 12-digit hexadecimal number, which is 48 bits in length. 

That is sourced directly from the hardware, the network interface card itself has this address burned into it and the address is globally unique and assigned by the manufacturer, so it's not very common to have manufacturers produce network cards with duplicate addresses but it has happened in the past because they think an address that was used quite some time ago is now defunct, so it's possible, but very unlikely. 

Generally, duplicate MAC addresses occur when the vendor-assigned address is not used, in other words, it's replaced with a self-assigned address or what's referred to as a locally administered address - again, this is not particularly common either but there are certain adapters that allow you to effectively change what MAC address is going to be used for that device. 

Can find out what address is being used with the ``ARP`` utility - can view or modify the ``ARP`` cache and the entries include the MAC addresses and it basically says that this MAC address is currently using this IP address and the ``ARP`` utility does allow you to view and/or modify those values, so you can use this to troubleshoot problems that are related to duplicate IP addresses or MAC addresses but again, in most cases, almost all network administrators are generally fine with using the burned in MAC address but if it is a case, for example, where a server has a faulty network adapter card, and you replace the adapter card, then you do have to assign the IP address that was being used for that system to the same adapter but now there is a different MAC address, so you do have to update some of the values, possibly, in some ``ARP`` caches to ensure that they are aware of the new MAC address using that old IP address. 

Again, that might not be an issue because ``ARP`` caches don't remain current for all that long - only about two minutes, by default, so if the cache is simply left to purge out, then ultimately it will update itself anyway but it can be an issue if you have these very important systems that need to be able to communicate via the MAC address and the IP address, so changes can result in problems but again, issues with duplicate MAC addresses are far less common than issues with duplicate IP addresses but always something to be on the lookout for.

***
###### DHCP Scope
***

Troubleshooting some ``DHCP`` issues, beginning with the scope itself.

This is something that you configure on the server and it represents the valid range of IP addresses that are available for assignment and any one address from that range is leased to the client computers for a certain amount of time and that is configurable as well.

The scopes are configured to designate a pool of IP addresses - the server can then allocate to the ``DHCP`` clients. 

The clients are required to be configured to use ``DHCP``, in other words, you have a choice, can set the client to use a static configuration or you can simply tell it to obtain an IP address automatically, in which case you are telling it to request its configuration from a ``DHCP`` server. 

Now this in fact is the default on Windows systems, but it can certainly be changed.

< [Table of Contents](#top) | [References](#references) >
<a id="DHCPScopeProperties"></a>
***
###### DHCP Scope Properties
***

With respect to the scope itself, there are several properties, these are only a few examples:


- does of course have to indicate the network ID and the subnet mask that you are going to use.


- the actual IP address range, the starting address through to the ending address.


- the lease duration and this is typically around a week by default for most systems but you can set it to really just about any value you want and it typically comes down to how many clients you have versus how many addresses you want to configure within a scope and the idea being that if you have only a few addresses but a lot of systems that maybe aren't all on at the same time, such as shift work, then you would typically have a fairly short lease duration, if you have the exact opposite, a lot of addresses and not that many clients, then any one client can keep that configuration for a fair amount of time, so again, the duration is your call but seven, eight days is usually around the default. 


- the name of the scope is simply what you can call it.


- an exclusion range is a set of addresses that falls within the scope, but that should not be assigned, so let's just say, your scope is ``192.168.1.100`` to ``200``, normally, it would assign out every one of those addresses from ``100`` to ``200`` but if you know, for example, that ``150`` to ``160`` is already being used because they've been statically assigned to maybe ten servers, we'll say then that represents the exclusion range, so again, it's really just only values that would normally fall within the range that you just don't want to assign.

<a id="TroubleshootingScopes"></a>
***
###### Troubleshooting Scopes
***

So when troubleshooting scopes, some common issues can include:

- the ``DHCP`` scope itself not being activated, this is generally an option you can configure when you are setting things up initially, for example, you might have many scopes that you need to create, but you don't have to activate them right then and there, as soon as it is activated, the server will start allocating those addresses, but maybe you just don't want it to do so yet - can leave them in a deactivated state until everything is fully configured but of course, if it was supposed to be activated and you just forgot, well then clearly that's a problem. 


- the lease configuration - anything wrong with the lease in terms of the duration and/or the validity periods - Maybe you just made a mistake there. 


- limited scope typically refers to a scope that does not have enough addresses to assign all of the clients, so you just made it too small basically. 


- the ``DHCP`` service itself might have been stopped, any given service on any given server could fail at any point, so if that service does fail, then that server will stop allocating addresses. 


- perhaps something has simply been misconfigured on the client side - the ``DHCP`` client service may have failed, maybe somebody set a particular client system to be statically configured and they weren't supposed to, something along those lines, so there can certainly be a lot of issues.

###### Rogue DHCP Server - MiTM
***

Might also encounter a rogue ``DHCP`` server, which is simply an unauthorized ``DHCP`` server appearing on a network, one that is not under administrative control. 

This is often used in man-in-the-middle attacks, because every version of Windows Server, for example, allows you to create a ``DHCP`` scope, so an attacker could very easily create a new ``DHCP`` server and then start handing out bogus IP address configurations to redirect clients. 

It offers the same services as an authorized ``DHCP`` server, so it will absolutely supply you with an IP address, a default gateway, and a DNS server but of course, if it is an unauthorized server doing so, that address configuration would be incorrect. 

< [Table of Contents](#top) | [References](#references) >

***
###### Detecting Rogue DHCP Server
***

Can try to detect rogue ``DHCP`` servers by using the ``DHCPLOC`` or locate utility - is a free download available from Microsoft, which is quite simply used to detect the presence of an unauthorized ``DHCP`` server.

Can also use the ``ipconfig /all`` command on any given client to confirm the IP address of the ``DHCP`` server because, of course, a rogue ``DHCP`` server would not have the same IP address as an authorized server. 

< [Table of Contents](#top) | [References](#references) >

***
###### Mitigating Rogue DHCP Server
***

So in terms of mitigating rogue ``DHCP`` servers, you can:


- implement intrusion detection systems - this does require the appropriate signature to detect ``DHCP`` packets. 


- use multi-layer switches which are capable of dropping packets that aren't authorized.


- implement ``DHCP`` snooping, which also drops ``DHCP`` messages that are not trusted.


If you are in a Windows - Active Directory environment - the Active Directory itself, authorizes ``DHCP`` servers but this, of course, would only be applicable to Windows/Active Directory client systems, they are then informed to not accept an IP address configuration from any server that is not authorized but again, that's only applicable to Windows and Active Directory environments - anything other than that would not be able to detect, of course, the presence of an Active Directory authorized ``DHCP`` server.

###### DHCP Lease Time
***

Looking at how time can affect the configuration of a system, particularly with respect to DHCP, as well as some other components. 

When it comes to your IP address configuration that you receive from a DHCP server, there is always what's known as a lease duration on that configuration, meaning that you have it for a certain amount of time before it has to be renewed and this is applicable to basically any type of device, whether it's a laptop, a desktop, a smartphone, even a smart TV these days, or a tablet. 

Those types of devices toward the bottom there, the smartphones, the smart TVs, and tablets, certainly much more common in a home environment, but it's the same idea, that any time you get an IP configuration from a DHCP server, there is this time associated with the lease.

In terms of the optimum length of what that time should be, it really is up to you, but in an environment such as a small home network, typically there are more addresses than devices, so in that case, you can have a longer lease period because it's unlikely that you will run out of addresses. 

Most home routers, for example, default to have about a 100 addresses available, and it's usually something in the ``192.168.``, let's just say ``1.100`` to ``200`` range - that's fairly common, so that represents 100 available addresses, so the chance that you would have more than 100 devices leasing IP addresses from that server in a home environment is fairly low. 

But if you are in a coffee shop, for example, then you might have many guests using your Wi-Fi, but only for a short period of time, so in this case, a shorter lease period would make much more sense, don't want someone to lease an address for, let's say, a week, when they are only going to be there for half an hour, so you can adjust the length of the lease quite a lot, but it ultimately does come down to how long you feel is appropriate for any given device to have the same address configuration. 

In a corporate environment, again, the default is usually around a week for most devices to have a valid address configuration, but again, if you have far more addresses than devices, you can look at increasing that validity period if you want to or if you have the exact opposite, more devices, not that many addresses, you can still shorten it - really comes down to whatever you feel is appropriate.

###### Expired IP Addresses
***

An expired IP address is essentially the fact that your lease does expire, so you can lose the network configuration for any given network. 

This might interfere with other devices or services if a device continues to use an IP address from an expired lease, so, again, the idea is if you did receive a particular configuration from a DHCP server and the lease says it's ``x`` amount of days, then, of course, at that point in time, when the expiry comes, that configuration is no longer valid, if your system is turned on and is able to contact the DHCP server, it will simply request a new configuration and you shouldn't have too much trouble at that point - but of course, you might lapse over the expiry time with your system turned off. 

Maybe you know, for example, that the lease configuration is going to expire, maybe let's say at the end of today, but if you shut down your system and then leave for a long weekend or something like that, then that address is still attached to your device the entire time it's off and then when you boot it up, let's say, maybe over the weekend if you're at home because it was a laptop, then it's possible it will still try to use that expired lease and it might interfere with trying to get an address in your home network or something along those lines. 

In the meantime, the server itself might attempt to lease the same IP address to another client on the network because it believes that address, since it did expire, in essence, is now available, so it might lease it out to someone else, so then you come back to work on Monday morning and you boot up your system and it still has that old address, then you might end up with an IP address conflict, so again, it depends on the duration and the state of the system when that expiry happens, so these aren't particularly common scenarios, but they can arise.

###### Incorrect System Times
***

The time on the system itself - might be incorrect - computer clocks can be inaccurate, and this applies to servers, workstations, and any other network device that maintains the time.

In some cases, they can still be configured manually and this effectively refers to the fact that the clock can be maintained by a battery, which can drift slightly on any given day, if the battery is getting low, for example, very often you will notice the time starting to lag behind a little bit, and that can have an effect because time synchronization is critical for logging, managing, securing, planning, and debugging and can also cause network issues, such as the inability to join an Active Directory domain, and various other synchronization problems and this comes down to the fact that clients, in essence, need to agree on what time it is as per their servers - the Kerberos protocol, for example, in a Windows Active Directory environment only has a tolerance of five minutes. 

If they do not agree that the time is within that five minutes - by default - then it will not let that client participate because it can throw off logging and managing and securing those systems. 

Particularly, when it comes to security, if the time is way off, then whatever it did log in terms of security just does not represent the correct time and that can also have an effect on your DHCP lease if the time is not correct.

###### Network Time Protocol - NTP | Public Internet Time Servers

This way you know that your logs, and your security settings, and your policies, and your DHCP leases are all operating on systems that agree on what time it is.

###### Blocked Ports
***

With respect to blocked ports, ports simply represent the endpoint of communication for specific applications and services.

There are ``1024`` well-known port numbers for services that are very common within most environments and most applications - used for both TCP or UDP.

This comes down to the method of transmission, again, TCP uses acknowledgements and is a little more reliable, but UDP is a little faster because it does not use acknowledgements, but each of them have ports for their specific applications, and you quite simply need to know which ports need to be open in terms of communication. 

Can use a port scanner to determine if any ports are open on a network and they can be used to scan a single port, a range of ports, or look for any vulnerable ports, in other words, ports that should not be open, but are.

With respect to whether or not a port should be open comes down to whether or not you are using that service or that application.

Quickly referring back to ``SSL``, for example, uses port ``443``, so, clearly, if you want your environment to be able to use SSL, then ``443`` needs to be open.

If you have any other service that you are not using, such as ``Telnet``, for example, then that port should be closed, that's port ``23``.

This, of course, just gives you a little extra security, because if the port is entirely closed, then an attacker cannot use that port to try to get into your environment, so by scanning for the vulnerable ports, those that are open but aren't being used, you can, of course, improve the security of your environment.

i. Describe why a duplicate IP address might occur. 

There can be a few different reasons, but most often this is due to the use of static IP addresses. If you are manually configuring systems with a preconfigured block of addresses, let's say, you're simply more likely to assign the same address twice. 

So if you are doing this, which is perfectly fine, by the way, you just need to be careful - make sure that there are no typos - make sure that it's documented with respect to which address is assigned to which system - also need to ensure that if you are using, let's just say, a block of static addresses, that they don't overlap at all with what is being assigned by a DHCP server.

Need to ensure that they, effectively, avoid each other, so that the DHCP server does not issue out an address that has already been statically assigned - either way, will end up with the same problem, the duplicate IP address. 

ii. Describe issues relating to an incorrect subnet mask. 

Quite simply, any kind of mistake in the subnet mask, in most cases, will prevent you from being able to connect to other clients entirely. 

Your system will likely believe that it's on a different subnet, so it won't be able to communicate with anyone - in short, you need to verify that all systems in the same subnet have the exact same subnet mask.

iii. Describe a rogue DHCP server and its effect on clients. 

The rogue is simply an unauthorized DHCP server, and it's often used in man-in-the-middle attacks to assign bogus IP address configurations to redirect clients to something like a malicious website. 


Basically, it's just a DHCP server that has appeared on the network without the knowledge or the consent of administration. 

It's fairly easy to do this if you have physical access to the network and since DHCP is broadcast-based, all of the clients, in essence, will see that DHCP server, so if they can get it to assign these addresses to those clients, then, in essence, they have an invalid configuration and, again, can be misused.

- can be set to Manual, which means you, as an administrator, you have to start it
